```python
import os
import gym
from stable_baselines3 import PPO
from stable_baselines3.common.vec_env import DummyVecEnv
from stable_baselines3.common.evaluation import evaluate_policy
```

```python
# Setup the environment for RL

environment_name = 'CartPole-v0'
env = gym.make(environment_name)
```

```python
# Test the invironment

episodes = 5
for episode in range(1, episodes + 1):
    state = env.reset()
    done = False
    score = 0
    
    while not done:
        env.render()
        action = env.action_space.sample()
        n_state, reward, done, info = env.step(action)
        score += reward
    print(f"Episode: {episode}  Score: {score}")
env.close()
```

```python
logs_path = os.path.join('Training', 'logs')
```

```python
env = gym.make(environment_name)
env = DummyVecEnv([lambda: env])
model = PPO('MlpPolicy', env, verbose=1, tensorboard_log=logs_path)
```

```python
# Train the RL model
model.learn(total_timesteps=100000)
```

```python
PPO_path = os.path.join('Training', 'Saved Models', 'PPO_MODEL_CARTPOLE')
```

```python
# Save the model
model.save(PPO_path)
```

```python
# Delete model for testing
del model
```

```python
# Load the model again
model = PPO.load(PPO_path, env=env)
```

```python
# Evaluate the model and see how it performs
evaluate_policy(model, env, n_eval_episodes=10, render=True)
```

```python
env.close()
```

```python
# Test the model

episodes = 5
for episode in range(1, episodes + 1):
    observations = env.reset()
    done = False
    score = 0
    
    while not done:
        env.render()
        action, _ = model.predict(observations) #Now using our model to make decisions
        observations, reward, done, info = env.step(action)
        score += reward
    print(f"Episode: {episode}  Score: {score}")
env.close()
```

```python
from stable_baselines3.common.callbacks import EvalCallback, StopTrainingOnRewardThreshold
```

```python
save_path = os.path.join('Training', 'Saved Models')
```

```python
# Adding stop callback on specific threshold
stop_callback = StopTrainingOnRewardThreshold(reward_threshold=200, verbose=1)
eval_callback = EvalCallback(env,
                            callback_on_new_best=stop_callback,
                            eval_freq=10000,
                            best_model_save_path=save_path,
                            verbose=1)
```

```python
model = PPO('MlpPolicy', env, verbose=1, tensorboard_log=logs_path)
```

```python
model.learn(total_timesteps= 50000, callback=eval_callback)
```

```python
# Changing architecture and policies if the model not performs well
net_arch = [dict(pi=[128, 128, 128, 128], vf=[128, 128, 128, 128])]
```

```python
model = PPO('MlpPolicy', env, verbose=1, tensorboard_log=logs_path, policy_kwargs={'net_arch': net_arch})
```

```python
model.learn(total_timesteps= 50000, callback=eval_callback)
```
